### Pandas Export

- to csv
- to excel
- to html

```python
import pandas as pd
df = pd.DataFrame()
```

## to_csv

```python
df = pd.read_csv("C:\\Users\\ACER\\Desktop\\DSMP1\\Datasets\\datasets-session-20\\deliveries.csv")
df.head()
```

```python
temp_df = df.groupby('batsman')['batsman_runs'].sum().reset_index()
```

```python
temp_df.to_csv("batsman_runs.csv", index=False)
```

```python
df.pivot_table(index='batsman', columns='bowling_team', values='batsman_runs', aggfunc='sum').to_csv('batsman_vs_teams.csv')
```

## to_excel

```python
df
```

```python
temp_df = df.groupby('batsman')['batsman_runs'].sum().reset_index()
```

```python
temp_df.to_excel("batsman_runs_ipl.xlsx")
```

```python
temp_df.to_excel("batsman_runs_ipl.xlsx", sheet_name='batsman_runs')
```

```python
temp_df2 = df.pivot_table(index='batsman', columns='bowling_team', values='batsman_runs', aggfunc='sum')
```

```python
with pd.ExcelWriter('output.xlsx') as writer:
    temp_df.to_excel(writer, sheet_name='Sheet_name_1')
    temp_df2.to_excel(writer, sheet_name='Sheet_name_2')
```

## to_html

```python
df.query('batsman_runs == 6').pivot_table(index='over', columns='ball', values='batsman_runs', aggfunc='count').to_html('sixes_heatmap.html')
```

## to_json

```python
df
```

```python
df.groupby(['batting_team','batsman'])['batsman_runs'].sum().unstack().to_json('ipl.json')
```
