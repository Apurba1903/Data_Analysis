```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
```

```python
# Configure Chrome options
chrome_options = Options()
chrome_options.add_argument("user-agent=Mozilla/5.0 (Windows NT 6.3; Win 64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.162 Safari/537.36")

# Initialize WebDriver
driver = webdriver.Chrome(options=chrome_options)
driver.maximize_window()

# Make the request
driver.get("https://www.ambitionbox.com/list-of-companies?page=1")

# Get page HTML
webpage = driver.page_source
```

```python
soup = BeautifulSoup(webpage, 'lxml')
```

```python
company = soup.find_all('div', class_="companyCardWrapper__primaryInformation")
```

```python
name = []
rating = []
review = []
companyType_and_location = []


for i in company:
    
    name.append(i.find('h2').text.strip())
    rating.append(i.find('div', class_='rating_text').text.strip())
    review.append(i.find('span', class_='companyCardWrapper__companyRatingCount').text.strip())
    companyType_and_location.append(i.find('span', class_='companyCardWrapper__interLinking').text.strip())
    
    
    
d = {'name':name, 'rating':rating, 'review':review, 'type&loc':companyType_and_location}

df = pd.DataFrame(d)

df
```

```python
final = pd.DataFrame()

for j in range(1,11):
    
    url = "https://www.ambitionbox.com/list-of-companies?page={}".format(j)
    
    chrome_options = Options()
    chrome_options.add_argument("user-agent=Mozilla/5.0 (Windows NT 6.3; Win 64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.162 Safari/537.36")

    # Initialize WebDriver
    driver = webdriver.Chrome(options=chrome_options)
    driver.maximize_window()

    # Make the request
    driver.get(url)

    # Get page HTML
    webpage = driver.page_source
    
    
    soup = BeautifulSoup(webpage, 'lxml')
    
    company = soup.find_all('div', class_="companyCardWrapper__primaryInformation")
    
    
    name = []
    rating = []
    review = []
    companyType_and_location = []


    for i in company:
        
        name.append(i.find('h2').text.strip())
        rating.append(i.find('div', class_='rating_text').text.strip())
        review.append(i.find('span', class_='companyCardWrapper__companyRatingCount').text.strip())
        companyType_and_location.append(i.find('span', class_='companyCardWrapper__interLinking').text.strip())
        
        
        
    d = {'name':name, 'rating':rating, 'review':review, 'type&loc':companyType_and_location}

    df = pd.DataFrame(d)
    
    final = pd.concat([final, df], ignore_index=True)
```
