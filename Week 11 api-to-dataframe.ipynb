```python
df = pd.DataFrame()
```

```python
for i in range(1,507):
    
    response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=68a51b69994c2f3740fb1fa9314bb9ea&language=en-US&page={}'.format(i))
    
    temp_df = pd.DataFrame(response.json()['results'])[['id','original_title','overview','release_date','popularity','vote_average','vote_count']]

    df = pd.concat([df, temp_df], ignore_index=True)
```
